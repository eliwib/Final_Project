My project is on the concentration of Chlorophyll in the Prince William Sound. The sound is considered an important area for oil, shipping, and wildlife so it's important to study. Chlorophyll helps to inform us about the amount of phytoplankton in the water. Phytoplankton is considered a "health" indicator for aquatic systems. Too much or too little means that the nutrients aren't properly proportioned. For this project I wanted to find if Chlorophyll cocentrations were dictated by latitude in the sound. My I had two hypothesis for the data; 1. There would be spikes in the data when we see spring nutrient upwelling, and 2. The Chloropyll concentration would be higher at higher latitudes due to increased nutrient upwelling.

## 1. Importing Data and Finding Best Measurement Depth 

<!-- In this cell, introduction the data set(s) you are working with, specify which aspect of your research question this section addresses, and describe the figure you will produce. -->

To start I had to import my Chloropyll data sets using the read excel function. I put all my data sets into one data frame using the .concat function then I began cleaning up all of my data. For the data from 2017-2020 the Chlorophyll data had one title, but the title changed for 2021. So I used .notnull to find the data values for Chloropyll and combinded them into the same column so I could plot all the values together. Then I began seperationg my data into depth profiles so I could find what depth had the highest average concentration of Chlorophyll. This would just make my comparisions easier to make. The bar plot shows the average chlorophyll concentration with the highest concentrations being at 5 and 10 meters of depth. While this doesn't helpt to answer my question, it does justify the analyzing choices I make for the rest of the data.

```python
# load data file(s)
chladf_17 = pd.read_excel('Chla_2017_190628.xlsx',index_col='Date',parse_dates=True)
chladf_18 = pd.read_excel('Chla_2018_190628.xlsx',index_col='Date',parse_dates=True)
chladf_19 = pd.read_excel('Chla_2019_200721.xlsx',index_col='Date',parse_dates=True)
chladf_20 = pd.read_excel('Chla_2020_211130.xlsx',index_col='Date',parse_dates=True)
chladf_21 = pd.read_excel('Chla_2021_221201.xlsx',index_col='Date',parse_dates=True)
```

```python
# perform data cleaning steps
#Combining all my dataframes into one data frame
raw_data_df=pd.concat([chladf_17,chladf_18,chladf_19,
                       chladf_20,chladf_21],axis=0)

#Combining the two Chlorophyll coulums into one
#Removing the Null Values from the two columns
df_1 = raw_data_df[raw_data_df['Chlorophyll-a (µg/l)'].notnull()]
df_2 = raw_data_df[raw_data_df['Chlorophyll-a (Âµg/l)'].notnull()]
#Putting the values in the same column
df_2['Chlorophyll-a (µg/l)']=df_2['Chlorophyll-a (Âµg/l)']
#Dropping the extra Chlorophyll column
df_3=df_1.drop(labels='Chlorophyll-a (Âµg/l)',axis=1)
df_4=df_2.drop(labels='Chlorophyll-a (Âµg/l)',axis=1)
#Making my new Dataframe
total_chla_data=pd.concat([df_3,df_4,],axis=0)
display(total_chla_data)
```

```python
#My Data analysis
#Finding the depth at which the most chlorophyll exists
#Slice each depth out of the dataframe
depth_0=total_chla_data[total_chla_data['Depth (m)']==0]
depth_5=total_chla_data[total_chla_data['Depth (m)']==5]
depth_10=total_chla_data[total_chla_data['Depth (m)']==10]
depth_15=total_chla_data[total_chla_data['Depth (m)']==15]
depth_25=total_chla_data[total_chla_data['Depth (m)']==25]
depth_40=total_chla_data[total_chla_data['Depth (m)']==40]
depth_50=total_chla_data[total_chla_data['Depth (m)']==50]

#Taking the mean Chloropyll amount at each depth
chla_0=depth_0['Chlorophyll-a (µg/l)'].mean()
chla_5=depth_5['Chlorophyll-a (µg/l)'].mean()
chla_10=depth_10['Chlorophyll-a (µg/l)'].mean()
chla_15=depth_15['Chlorophyll-a (µg/l)'].mean()
chla_25=depth_25['Chlorophyll-a (µg/l)'].mean()
chla_40=depth_40['Chlorophyll-a (µg/l)'].mean()
chla_50=depth_50['Chlorophyll-a (µg/l)'].mean()
print(chla_0)

#Finding the Standard error of each
err_0=depth_0['Chlorophyll-a (µg/l)'].sem()
err_5=depth_5['Chlorophyll-a (µg/l)'].sem()
err_10=depth_10['Chlorophyll-a (µg/l)'].sem()
err_15=depth_15['Chlorophyll-a (µg/l)'].sem()
err_25=depth_25['Chlorophyll-a (µg/l)'].sem()
err_40=depth_40['Chlorophyll-a (µg/l)'].sem()
err_50=depth_50['Chlorophyll-a (µg/l)'].sem()
print(err_0)
```

```python
#Ploting the first figure
#making our tuples for our bar chart
depth_nums=(0,5,10,15,25,40,50)
x_pos=np.arange(len(depth_nums))
y_vals=(chla_0,chla_5,chla_10,chla_15,
        chla_25,chla_40,chla_50)
error=(err_0,err_5,err_10,err_15,err_25,
       err_40,err_50)
#Making our bar chart
fig,ax=plt.subplots(figsize=(10,5))
ax.bar(x_pos,y_vals,yerr=error,color='seagreen')
ax.set_xticks(x_pos,labels=depth_nums)
ax.set_xlabel('Depth (m)')
ax.set_ylabel('Chlorophyll-a (µg/l)')
ax.set_title('Depth of Mean Chlorophyll-a')
ax.grid(axis='y',linestyle='--')

#Add in error labales with two decimal places (The f string does the rounding)
for i in range(len(x_pos)):
    plt.text(x_pos[i], y_vals[i] + error[i] + 0.012, f'{y_vals[i]:.2f}', ha='center')
plt.savefig('Mean_Chlorophyll_Depths')
```

<!-- Describe how this figure helps address your research question. What patterns or insights can you observe? -->
**Figure 1: Bar graph that displays the average Chlorophyll Concentrations at each measured depth. Depths of 5 and 10 meters had the highest concentrations. I choose to do my analysis at 10 meters of depth since it had a high concentration which suggested more data to work with.**

## 4. Average Chlorophyll Cocentrations

<!-- In the introduction below, specify the data set(s) you are working with in this section, which aspect of your research question this section addresses, and the figure you will produce -->

Since the previous graph was a little hard to read I took the average Chloropyll concentrations at the two stations. This would help tell me which station had a higher or lower Chlorophyll concentration.

```python
# perform data analysis
#Getting the mean chlorophyll for each station
mean_chla_MS = station_7['Chlorophyll-a (µg/l)'].mean()
mean_chla_EH = station_11['Chlorophyll-a (µg/l)'].mean()
print(mean_chla_EH,mean_chla_MS)
#Getting the standard error for each station
err_MS = station_7['Chlorophyll-a (µg/l)'].sem()
err_EH = station_11['Chlorophyll-a (µg/l)'].sem()
print(err_EH,err_MS)
```

```python
# make and display figure
Stations=('EH','MS')
x_pos=np.arange(len(Stations))
y_vals=(mean_chla_EH,mean_chla_MS)
error=(err_EH,err_MS)
#Making our bar chart
fig,ax=plt.subplots(figsize=(5,5))
ax.bar(x_pos,y_vals,yerr=error,color='seagreen')
ax.set_xticks(x_pos,labels=Stations)
ax.set_xlabel('Station')
ax.set_ylabel('Chlorophyll-a (µg/l)')
ax.set_title('Mean Chlorophyll')
ax.grid(axis='y',linestyle='--')

plt.savefig('MS_EH_Chla')
```

<!-- Describe how this figure helps address your research question. What patterns or insights can you observe? -->
**Figure 4:** This graph shows that station MS has over double the amount of chlorophyll found at Station EH. This is the opposite of what my hypothesis stated, meaning that the lower latitude station had a higher concentration of Chlorophyll than the higher latitude station.

## 5. Chlorophyll Concentrations around the Sound

<!-- In the introduction below, specify the data set(s) you are working with in this section, which aspect of your research question this section addresses, and the figure you will produce -->
The earlier graphs told me that Chlorophyll was more abundant at lower latitudes, but I decided to see if this trend was the same for other stations around the Prince William sound. So I took a couple other stations and graphed their average Chlorophyll Concentrations to compare them. The resulting graph could help me understand if latitude was the driving factor between the Chlorophyll Concentrations.

```python
#Getting the mean chlorophyll for each station
mean_chla_MS = station_7['Chlorophyll-a (µg/l)'].mean()
mean_chla_EH = station_11['Chlorophyll-a (µg/l)'].mean()
mean_chla_SM = station_2['Chlorophyll-a (µg/l)'].mean()
mean_chla_HE = station_3['Chlorophyll-a (µg/l)'].mean()
mean_chla_WM = station_9['Chlorophyll-a (µg/l)'].mean()
#Getting the standard error for each station
err_MS = station_7['Chlorophyll-a (µg/l)'].sem()
err_EH = station_11['Chlorophyll-a (µg/l)'].sem()
err_SM = station_2['Chlorophyll-a (µg/l)'].sem()
err_HE = station_3['Chlorophyll-a (µg/l)'].sem()
err_WM = station_9['Chlorophyll-a (µg/l)'].sem()
```

```python
#Making my lists for the figure
Stations=('EH','HE','MS','SM','WM')
x_pos=np.arange(len(Stations))
y_vals=(mean_chla_EH,mean_chla_HE,mean_chla_MS,
       mean_chla_SM,mean_chla_WM)
error=(err_EH,err_HE,err_MS,err_SM,err_WM)
#Making my bar graph
fig,ax = plt.subplots(figsize=(10,5))
ax.bar(x_pos,y_vals,yerr=error,color='seagreen')
ax.set_xticks(x_pos,labels=Stations)
ax.set_xlabel('Station')
ax.set_ylabel('Chlorophyll-a (µg/l)')
ax.set_title('Mean Chlorophyll')
ax.grid(axis='y',linestyle='--')
```

<!-- Describe how this figure helps address your research question. What patterns or insights can you observe? -->
**Figure 5:** The figure has the mean chlorophyll concentrations of a couple of stations around the Prince William Sound. Station WM has a higher concentration than MS, and SM has more than HE, which suggests that Latitude isn't what drives the concentration of Chlorophyll in the Prince William Sound.
